```
import pandas as pd
```

```
sht1 = gc.open_by_url('https://docs.google.com/spreadsheets/d/1ygU1bcKRam2NJE1ru6xXgY6A7IqNjz4Qjhk8qyG9jlE/edit#gid=75300064')
worksheet = sht1.worksheet("Sheet1")
```

```
df = pd.DataFrame(worksheet.get_all_records())
```

```
import plotly.graph_objects as go
```

```
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = ["A1", "A2", "B1", "B2"], #, "C1", "C2"],
      color = "blue"
    ),
    link = dict(
      source = [0, 1, 0, 2],#, 3, 3], # indices correspond to labels, eg A1, A2, A2, B1, ...
      target = [2, 3, 3, 4],#, 4, 5],
      value = [8, 4, 2, 8],#, 4, 2]
  ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.show()
```

```
dic = {item:index for index, item in enumerate(df['course'].unique().tolist())}
df['course_'] = df['course']
df.replace({"course_": dic}, inplace=True)
```

```
dic_ = {item:index for index, item in enumerate(df['Адрес электронной почты'].unique().tolist())}
df['email'] = df['Адрес электронной почты']
df.replace({"email": dic_}, inplace=True)
```

```
label_ = df['Адрес электронной почты'].tolist()
source_ = df['email'].tolist()
target_ = df['course_'].tolist()
value_ = [1 for i in range(len(df['course'].tolist()))]
```

```
label_1 = df['Адрес электронной почты'].tolist()
label_2 = df['course'].tolist()

label = label_1 + label_2
source_ = df['email'].tolist()

target_ = df['course_'].tolist()
target__ = [i+407 for i in target_]
value_ = [1 for i in range(len(df['course'].tolist()))]
```

```
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = label, # здесь баг - надо написать функцию, которая будет выдавать индекс таргета (таргет надо сделать без повторений) (параметры фигуры вместо df откатить к листам)
      color = "blue"
    ),
    link = dict(
      source = source_,
      target = target__,
      value = value_,
  ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.update_layout(width=1500, height=2000)
fig.show()
```

```
df_ = pd.DataFrame(list(zip(label, source_, target__, value_)), columns =['label', 'source', 'target', 'value'])
```

```
fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = df_['label'].tolist(),
      color = "blue"
    ),
    link = dict(
      source = df_['target'].tolist(),
      target = df_['source'].tolist(),
      value = df_['value'].tolist(),
  ))])

fig.update_layout(title_text="Basic Sankey Diagram", font_size=10)
fig.update_layout(width=1500, height=2000)
fig.show()
```
